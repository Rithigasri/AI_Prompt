```python
import requests
import json
import time  # Import time module to add delays

# Your Gemini API key
api_key = "AIzaSyCS-maWiatQZM3um73Dse4Gj9vx4_fpAbg" #Enter your API Key

# Gemini API endpoint
url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent?key={api_key}"


# Read the content of the prompt.txt file
with open("prompt1.txt", "r", encoding="utf-8") as file:
    prompt_content = file.read()

# Function to interact with the user and gather information
def gather_information(prompt_content):
    time.sleep(1)  # Add a short delay before asking the first question

    while True:
        # Construct the request payload with the prompt content
        payload = {
            "contents": [
                {
                    "parts": [
                        {"text": prompt_content}
                    ]
                }
            ]
        }

        # Set the headers
        headers = {
            "Content-Type": "application/json",
        }

        # Make the API request
        response = requests.post(url, headers=headers, data=json.dumps(payload))

        # Check for a successful response
        if response.status_code == 200:
            result = response.json()
            question = result["candidates"][0]["content"]["parts"][0]["text"]

            # Display only the AI's question
            print(f"AI: {question}")
            time.sleep(1)  # Add delay before accepting user input

            user_input = input("Your answer (type 'exit' to stop): ").strip()

            # Print only the user's response
            print(f"\nUser: {user_input}\n")

            # Check if the user wants to exit
            if user_input.lower() == "exit":
                print("Exiting... Goodbye!\n")
                break

            # Update prompt_content to include the user's response, so the AI can generate more context-based questions
            prompt_content += f"\nUser: {user_input}"

            # Add a short pause before generating the next question
            time.sleep(1)

        else:
            print(f"Request failed with status code {response.status_code}: {response.text}")
            break

# Start the information gathering
gather_information(prompt_content)
```
